# Data Cleaning

```python
#drop the blank columns from the dataset
data.drop(['New','PendingS'],axis = 1, inplace = True)
```

```python
# check the null values From the Dataset.
data.isnull().sum()
```

```python
#drop the Null Values From the Dataset.
data.dropna(inplace = True)
```

# Change The Datatype of the Column

```python
#change The Data Type From Float to integer.
data['ship-postal-code'] = data['ship-postal-code'].astype('int')
```

```python
#Change the Date COlumn DataType From Object To Date_time.
data['Date'] = pd.to_datetime (data['Date'])
```

```python
#Change the Column Name 'Qty' To 'Quantity'.
data.rename(columns={'Qty':'Quantity'},inplace = True)
```

# Size

```python



#Chacking the B2B Data by using Pie Chart
B2B_check = data['B2B'].value_counts()

#plot the pie charts
plt.pie(B2B_check,labels=B2B_check,autopct='%1.1f%%')
#plt.axis('equal')
plt.show()
```

```python
#Chacking the B2B Data by using Pie Chart
B2B_check = data['B2B'].value_counts()

#plot the pie charts
plt.pie(B2B_check,labels=B2B_check.index,autopct='%1.1f%%')
#plt.axis('equal')
plt.show()
```

# Note : From the Above Chart we can see that maximum i.e.99.2% of Buyers are retailers and 0.8% are B2B buyers

# size of the Categorys

```python
# prepare Data for Scatter plot
x_data = data['Category']
y_data = data['Size']

#plot the Scatter plot
plt.scatter(x_data,y_data)
plt.xlabel('Category')
plt.ylabel('Size')
plt.title('Scatter Plot')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=data, x='ship-state')
plt.xlabel('ship-state')
plt.ylabel('Count')
plt.title('Distribution of States')
plt.xticks(rotation=90)
plt.show()
```

# Top 10 States

```python
# top 10 States
top_10_States = data['ship-state'].value_counts().head(10)

plt.figure(figsize=(12, 6))
sns.countplot(data=data[data['ship-state'].isin(top_10_States.index)], x='ship-state')
plt.xlabel('Ship-state',size=15)
plt.ylabel('Count')
plt.title('Distribution of States')
plt.xticks(rotation=70)
plt.show()
```

# Note : From the Above Graph you can see that most of the Buyers are MAHARASTRA State.

# Top 10 Citys

```python
#top 10 citys
top_10_city = data['ship-city'].value_counts().head(10)

#plot the count plot
plt.figure(figsize=(12, 6))
sns.countplot(data=data[data['ship-city'].isin(top_10_city.index)], x='ship-city')
plt.xlabel('Ship-city',size=15)
plt.ylabel('Count')
plt.title('Distribution of City')
plt.xticks(rotation=70)
plt.show()
```

# Note : From the Above Graph you can see that most of the Buyers From Bengaluru City.
